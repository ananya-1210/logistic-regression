The goal is to predict whether a user will make a purchase based on their browsing behavior and other attributes. # **10. Categorical AND Continous split**
---

```python
X_train_cat = X_train.select_dtypes(include = 'object')
X_train_con = X_train.select_dtypes(exclude = 'object')
```

```python
X_test_cat = X_test.select_dtypes(include = 'object')
X_test_con = X_test.select_dtypes(exclude = 'object')
```

---
<a name = Section12></a>
# **11. Scaling**
---

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train_con)
```

```python
X_train_cont_s = scaler.transform(X_train_con)
X_test_cont_s = scaler.transform(X_test_con)
```

```python
X_train_cont_S_DF = pd.DataFrame(X_train_cont_s, 
                                 columns = X_train_con.columns + '_S', 
                                 index = X_train_con.index)

X_test_cont_S_DF = pd.DataFrame(X_test_cont_s, 
                                 columns = X_train_con.columns + '_S', 
                                 index = X_test_con.index)
```

---
<a name = Section13></a>
# **12. Encoding**
---

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output = False)
ohe.fit(X_train_cat)
```

```python
X_train_cat_ohe = pd.DataFrame(data = ohe.transform(X_train_cat),
                               index = X_train_cat.index,
                               columns = ohe.get_feature_names_out())

X_test_cat_ohe = pd.DataFrame(data = ohe.transform(X_test_cat),
                               index = X_test_cat.index,
                               columns = ohe.get_feature_names_out())
```

# 13.Concatinating

```python
X_train_comb = pd.concat([X_train_cat_ohe, X_train_cont_S_DF], axis = 1 )
X_test_comb = pd.concat([X_test_cat_ohe, X_test_cont_S_DF], axis = 1 )
```

# 8.1. Accuracy

Accuracy simply measures how often the classifier correctly predicts. We can define accuracy as the ratio of the number of correct predictions and the total number of predictions.

# 8.2. Confusion Matrix



Confusion Matrix is a performance measurement for the machine learning classification problems where the output can be two or more classes. It is a table with combinations of predicted and actual values.
It is a table used to evaluate the performance of a classification model. It compares the actual values (true labels) with the predicted values from the model. The matrix consists of four key components:


    

                            Predicted No             predicted yes 

                   
Actual No (TN)           True Negative (TN)	      False Positive (FP)


Actual Yes (TP)	         False Negative (FN)	      True Positive (TP)

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
confusion_matrix = confusion_matrix(Y_test_pred,Y_test)

print(confusion_matrix)
```

```python
138/362
```

# Recall¶

The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. The recall is intuitively the ability of the classifier to find all the positive samples.

```python
from sklearn.metrics import recall_score
recall_score(Y_test,Y_test_pred)
```

# 8.3. Precision¶

Precision explains how many of the correctly predicted cases actually turned out to be positive. Precision is useful in the cases where False Positive is a higher concern than False Negatives. The importance of Precision is in music or video recommendation systems, e-commerce websites, etc. where wrong results could lead to customer churn and this could be harmful to the business.

```python
from sklearn.metrics import precision_score
precision_score(Y_test,Y_test_pred)
```

# 8.5. F-1 score¶

The F1 score can be interpreted as a harmonic mean of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. The relative contribution of precision and recall to the F1 score are equal.

```python
from sklearn.metrics import f1_score
f1_score(Y_test,Y_test_pred)
```

# 8.6. F-beta score

The F-beta score is the weighted harmonic mean of precision and recall, reaching its optimal value at 1 and its worst value at 0

```python
from sklearn.metrics import fbeta_score
fbeta_score(Y_test,Y_test_pred,beta=1.2)
```
